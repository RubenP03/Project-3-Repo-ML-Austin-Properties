```python
df = pd.read_csv(os.path.join("CoStar Zip Code Info Ruben.csv"))
df.head()
```

```python
target = df["Avg Asking Rent/Unit"]
target_names = ["Avg Rent"]
```

```python
data = df.drop("Sub-Market", axis=1)
features = data.columns
data.head
```

```python
#drop rows that all values are NA in the row
data.dropna(how="all",axis=0,inplace=False)
#drop columns that all values are NA in the column
data.dropna(how="all",axis=1,inplace=False)
```

```python
data['Address'] = data['Address'].astype('str')
```

```python
foo = data.fillna(value=0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(foo, target, test_size=.3, random_state=42)
```

```python
LinearRegression()
```

```python
#start model
lr = LinearRegression(n_jobs=3)
lr.fit(X_train,y_train)
```

```python
# check is score
lr.score(X_test,y_test)

#see the predictions
lr.predict(X_test)
```

```python
# Loop through different k values to see which has the highest accuracy
# Note: We only use odd numbers because we don't want any ties
train_scores = []
test_scores = []
for k in range(1, 755, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_score = knn.score(X_train, y_train)
    test_score = knn.score(X_test, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
    
plt.plot(range(1, 20, 2), train_scores, marker='o')
plt.plot(range(1, 20, 2), test_scores, marker="x")
plt.xlabel("k neighbors")
plt.ylabel("Testing accuracy Score")
plt.show()
```
